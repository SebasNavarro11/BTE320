````python
interest = 3 # interest rate (absolute value %)
principal = 3000
n = 1 # one year interest

final = principal * (1 + interest/100) ** n

print(final)
````

Different functions of "print"
print(object(s)), like the one used above to display a process.

````python
print(interest, principal)
print(final)
````

Print(sep=" "), separator, shows the two values separated

````python
"\n"
````

Print(end="\n") new line character. Controls what happens at the end of the displayed message. This special character separates the outputs into two different lines, shown in the second process


````python
print("Interest rate:", interest, sep="=>")
print("Number of years:", n)
````

````python
print("Interest rate:", interest, end="---")
print("Number of years:", n)
````

````python
print("Interest rate:", interest, sep=">>>", end="$$$")
print("Number of years:", n)
````

````python
print("Interest rate:", interest, "; Number of year:", n)
````

```python
"....{}...{}...".format(a,b)
{} acts as a placeholder within a string (dynamic string)

````python
print("Interest rate: {}; Number of years: {}" .format(interest, n))
````

```python
f"....{a}....{b}.."
f strings, same thing as dynamic strings but without all the extra formtting, just put an 'f' at the beginning

````python
print(f"Interest rate: {interest}; Number of years: {n}")
````
