```python
sales = float(input("Enter sales ($): "))
rate = float(input("Enter tax rate (%): "))

cogs = (sales * 0.4)
gross = sales - cogs
taxes = rate / 100 * gross
net = gross - taxes

print(f"Sales: ${sales}")
print(f"Cost of Good Sold: ${cogs}")
print(f"Gross Profit: ${gross}")
print(f"Taxes: ${taxes}")
print(f"Net income: ${net}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")
```
