```python
def summ(n):
  total = 0
  for i in range(1, n+1):
    total += i
  return total

def summR(n):
  if n == 1:
    return 1 # Base case
  else:
    return n + summR(n-1)

summ(10) == summR(10)
```

```python
def fib(n):
  if n == 0 or n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

fib(5)
```

```python
def quicksort(L):
  if len(L) < 2:
    return L
  else:
    pivot = L[len(L) // 2]
    smaller, equal, larger = [], [], []

    for item in L:
      if item < pivot:
        smaller.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        larger.append(item)

    return quicksort(smaller) + equal + quicksort(larger)

quicksort([8,2,5,1,6,7,0])
```
