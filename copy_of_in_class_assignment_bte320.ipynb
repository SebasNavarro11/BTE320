```python
sales = float(input("Enter sales ($): "))
rate = float(input("Enter tax rate (%): "))

cogs = (sales * 0.4)
gross = sales - cogs
taxes = rate * gross
net = gross - taxes

print(f"Sales: ${sales}")
print(f"Cost of Good Sold: ${cogs}")
print(f"Gross Profit: ${gross}")
print(f"Taxes: ${taxes}")
print(f"Net income: ${net}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")
```

```python
numbers = []

while True:
  num = float(input("Enter a number: "))
  if num == 0:
    break
  numbers.append(num)
numbers.sort(reverse=True)
print(numbers)
```

```python
product = {
    "P-01":9.95,
    "P-02":10.50
}

id = input("Enter product ID: ")

if id in product:
  print(product[id])
else:
  print(f"Product {id} not found...")
```

```python
product = {
    "P-01":9.95,
    "P-02":10.50
}

id = input("Enter product ID: ")

print(product.get(id, f"Product {id} not found..."))
```

```python
names = {}

while True:
  name = input("Give a name: ")
  if name == "":
    break

  if name in names:
    names[name] += 1
  else:
    names[name] = 1

print(names)
```
