```python
sales = float(input("Enter sales ($): "))
rate = float(input("Enter tax rate (%): "))

cogs = (sales * 0.4)
gross = sales - cogs
taxes = rate * gross
net = gross - taxes

print(f"Sales: ${sales}")
print(f"Cost of Good Sold: ${cogs}")
print(f"Gross Profit: ${gross}")
print(f"Taxes: ${taxes}")
print(f"Net income: ${net}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")
```

```python
import random

play = input("Do you want to play? (yes/no): ")

while play == "yes":
    user = input("Choose rock, paper, or scissors: ")
    computer = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer)

    if user == computer:
        print("It's a tie!")
    elif user == "rock" and computer == "scissors":
        print("You win!")
    elif user == "paper" and computer == "rock":
        print("You win!")
    elif user == "scissors" and computer == "paper":
        print("You win!")
    elif user in ["rock", "paper", "scissors"]:
        print("Computer wins!")
    else:
        print("Invalid choice")

    play = input("Do you want to play again? (yes/no): ")
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #return s[-1] + reverse(s[:-1]) #Take the last char and put it first
    #return reverse(s[1:]) + s[0] #Take the first character and put it last
    return s[-1] + reverse(s[1:-1]) + s[0]

reverse('12345')
```

```python
def count_names(n):
    counts = {}

    for i in range(n):
        name = input("Enter a name: ")

        if name in counts:
            counts[name] += 1
        else:
            counts[name] = 1

    return counts


result = count_names(5)
print(result)
```

```python
def roi(cost, revenue):
  assert isinstance(cost, (int, float)), "Cost must be numerical"
  assert isinstance(revenue, (int, float)), "Revenue must be numerical"
  assert cost != 0, "Revenue must not equal 0"

  return (revenue - cost) / cost

def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Principal must be numerical"
  assert isinstance(annual_rate, (int, float)), "Annual rate must be numerical"
  assert isinstance(months, (int, float)), "Months must be numerical"

  if principal <= 0 or months <= 0 or annual_rate <= 0:
  raise ValueError("Invalid Input Value")

  try:
    r = annual_rate / 12 / 100
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    raise Exception(f"Error: {e}")

  return pmt

```
