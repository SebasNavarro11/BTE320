```python
def roi(cost, revenue):
  assert isinstance(cost, (int, float)), "Cost must be numerical"
  assert isinstance(revenue, (int, float)), "Revenue must be numerical"
  assert cost != 0, "Revenue must not equal 0"

  return (revenue - cost) / cost

def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Principal must be numerical"
  assert isinstance(annual_rate, (int, float)), "Annual rate must be numerical"
  assert isinstance(months, (int, float)), "Months must be numerical"

  if principal <= 0 or months <= 0 or annual_rate <= 0:
   raise ValueError("Invalid Input Value")

  try:
    r = annual_rate / 12 / 100
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    raise Exception(f"Error: {e}")

  return pmt

```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, new_name):
    self.__name = new_name

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print(f"Student {self.__name} has already enrolled in course {course}.")
    else:
      self.__courses.append(course)
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    dist = (dx + dy) ** 0.5
    return dist

class Rectangle:
  def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4):
    self.p1 = Point(x1, y1)
    self.p2 = Point(x2, y2)
    self.p3 = Point(x3, y3)
    self.p4 = Point(x4, y4)

    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height

```

```python
square = Rectangle(0,3,3,3,0,0,3,0)
square.area()
```

```python
class Employee:
  def __init__(self, name, wage, hours_worked):
    self.name = name
    self.__wage = wage
    self.hours_worked = hours_worked

  def get_wage(self):
    return self.__wage

  def calculate_pay(self):
    return self.__wage * self.hours_worked

  def __str__(self):
    return f"Employee name: {self.name}\nHourly Charge: ${self.__wage}"

  def set_wage(self, new_wage):
    assert new_wage > 0
    self.__wage = new_wage

class Manager(Employee):
  def __init__(self, name, wage, hours_worked, bonus):
    super().__init__(name, wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours_worked

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary: ${employee.calculate_pay()}")
print(f"Manager {manager.name} weekly salary: ${manager.calculate_pay()} that includes a {manager.bonus * 100}%")
```

```python
import random

def monte_carlo_pi(N):
  acc = 0
  for i in range(N):
    x = random.random()
    y = random.random()
    if x ** 2 + y ** 2 <= 1.0:
      acc +=1
  return 4 * acc / N

%timeit monte_carlo_pi(1000000)
```
